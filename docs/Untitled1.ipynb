```python
import glob
import os
import re

folder_path = "/home/dmculve/workspace/data/imdb/aclImdb/test/neg"
save_path = "/home/dmculve/workspace/data/imdb/aclImdb/test/clean_neg/"
file_list = glob.glob(folder_path + "/*.txt")

for i in range(0, len(file_list)):
    file_path  = file_list[i]
    file_name = os.path.basename(file_path)
    f = open(file_path, "r")
    text = f.read()
    clean = re.sub("<br />", " ", text)
#     clean = re.sub(r'[^\w\s]', '', clean)
    clean = re.sub(",", " ", clean)
    clean = re.sub("\"", "", clean)
    
    clean_path = save_path + file_name
    print(clean_path)
    
    with open(clean_path, 'w', encoding='utf-8') as f:
        f.write(clean)

    print(clean)
```

```python
import glob
import logging
import os
import pandas as pd

small_path = '/home/dmculve/workspace/data/imdb/aclImdb/test/small_clean_neg/'
file_list = glob.glob(small_path + "/*.txt")

# res = pd.DataFrame(columns=['filename','content'])
res = pd.DataFrame()
for i in range (0,len(file_list)):
    file_path = file_list[i]
    file_name = os.path.basename(file_path)
    content = open(file_path, "r").read()    
    res = pd.concat([res, pd.DataFrame(data = {"filename" : [file_name], "contents" : [content]})])



```

```python
import json
import stanza

ner_pipeline = stanza.Pipeline(lang='english', processors='tokenize,ner')

def stanza_ner(content):
    doc = ner_pipeline(content)
    ners = []
    for ent in doc.entities:
        print(f'{ent.text}\t{ent.type}') 
        entry=(ent.text,ent.type)
        ners.append(entry)        
    return json.dumps(dict(ners))


res['ner'] = res['contents'].apply(stanza_ner)
```

```python
import spacy

SPACY_MODEL = None
def get_spacey_model():
    global SPACY_MODEL
    if not SPACY_MODEL:
        SPACY_MODEL = spacy.load("en_core_web_sm")
    return SPACY_MODEL

def spacy_ner(content):
    spacy_model = get_spacey_model()
    doc = spacy_model(content)
    
    ners = []
    for ent in doc.ents:
        print(f'{ent.text}\t{ent.label_}') 
        entry=(ent.text,ent.label_)
        ners.append(entry)        
    return json.dumps(dict(ners))

res['ner_spacy'] = res['contents'].apply(spacy_ner)
```
